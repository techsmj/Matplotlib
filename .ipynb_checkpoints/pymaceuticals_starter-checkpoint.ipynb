```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset

pymace_df= pd.merge(mouse_df, clinical_df, on= "Mouse ID")
# Display the data table for preview
pymace_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
Tumor_group = pymace_df.groupby(["Drug", "Timepoint"]).agg({"Tumor Volume (mm3)":"mean"})
# Convert to DataFrame
Tumor_group.apply(pd.Series)
# Preview DataFrame
Tumor_group.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
Tumor_stande = pymace_df.groupby(["Drug", "Timepoint"]).agg({"Tumor Volume (mm3)":"sem"})
# Convert to DataFrame
Tumor_stande.apply(pd.Series)
# Preview DataFrame
Tumor_stande.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
Tumor_groupdf= Tumor_group.pivot_table(index=['Timepoint'], columns='Drug', values='Tumor Volume (mm3)', fill_value='').rename_axis(None, axis=1).reset_index()
        
Tumor_groupdf1= Tumor_groupdf.set_index('Timepoint')

# Preview that Reformatting worked
Tumor_groupdf1.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
Tumor_standedf= Tumor_stande.pivot_table(index=['Timepoint'], columns='Drug', values='Tumor Volume (mm3)', fill_value='').rename_axis(None, axis=1).reset_index()
        
Tumor_standedf1= Tumor_standedf.set_index('Timepoint')

# Preview that Reformatting worked
Tumor_standedf1.head()
```

```python
Tumor_plot= Tumor_groupdf1.drop(columns = ["Zoniferol", "Stelasyn", "Ramicane", "Propriva", "Ceftamin", "Naftisol"])
Tumor_plot
```

```python
Tumor_error= Tumor_standedf1.drop(columns = ["Zoniferol", "Stelasyn", "Ramicane", "Propriva", "Ceftamin", "Naftisol"])
Tumor_error
```

```python

#plot error bars
ax = Tumor_plot.plt.errorbar(figsize=(12,8), yerr = Tumor_error, legend = False)
#reset color cycle so that the marker colors match
ax.set_prop_cycle(None)
#plot the markers
Tumor_plot.plt.errorbar(figsize=(12,8), style=['^-', 'o--', 'x-.', 'D-'], markersize=14, ax = ax)

plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.axis('tight')

plt.show()

# Save the Figure

#df.plot(figsize=(12, 8), 
       #style=['^-', 'o--', 'x-.', 'D-'], 
        #yerr=errors,
        #markersize=14)
```

```python
Met_plot= Met_groupdf1.drop(columns = ["Zoniferol", "Stelasyn", "Ramicane", "Propriva", "Ceftamin", "Naftisol"])
Met_plot
```

```python
# Minor Data Munging to Re-Format the Data Frames
Met_standedf= Met_stande.pivot_table(index=['Timepoint'], columns='Drug', values='Metastatic Sites', fill_value='').rename_axis(None, axis=1).reset_index()
        
Met_standedf1= Met_standedf.set_index('Timepoint')

# Preview that Reformatting worked
Met_standedf1.head()
```

```python
Met_error= Met_standedf1.drop(columns = ["Zoniferol", "Stelasyn", "Ramicane", "Propriva", "Ceftamin", "Naftisol"])
Met_error
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric) 
Mouse_ID_group = pymace_df.groupby(["Drug", "Timepoint"]).agg({"Mouse ID":"count"})
# Convert to DataFrame
Mouse_ID_group.apply(pd.Series)
# Preview DataFrame
Mouse_ID_group.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
Mouse_ID_groupdf= Mouse_ID_group.pivot_table(index=['Timepoint'], columns='Drug', values='Mouse ID', fill_value='').rename_axis(None, axis=1).reset_index()
        
Mouse_ID_groupdf1= Mouse_ID_groupdf.set_index('Timepoint')

# Preview that Reformatting worked
Mouse_ID_groupdf1.head()
```

```python
Mouse_IDdf= Mouse_ID_groupdf1.drop(columns = ["Zoniferol", "Stelasyn", "Ramicane", "Propriva", "Ceftamin", "Naftisol"])
Mouse_IDdf
```

```python
# Use numpy to create quartiles
#q1 =  (100. * Mouse_IDdf / Mouse_IDdf.sum()).round(0)
QMouse_IDdf= Mouse_IDdf.quantile([], axis = 0) 
QMouse_IDdf
```

```python
# Generate the Plot (Accounting for percentages)

#plot error bars
ax = QMouse_IDdf.plot(figsize=(12,8), legend = False)
#reset color cycle so that the marker colors match
ax.set_prop_cycle(None)
#plot the markers
QMouse_IDdf.plot(figsize=(12,8), style=['^-', 'o--', 'x-.', 'D-'], markersize=14, ax = ax)

plt.show()
```
